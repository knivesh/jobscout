```python
import requests
import json
from datetime import datetime
from pathlib import Path
```

```python
TOKEN = "adyen"
URL = f"https://boards-api.greenhouse.io/v1/boards/{TOKEN}/jobs"
PARAMS = {"content": "true"}   # include the full job description in each job
```

```python
r = requests.get(URL, params=PARAMS, timeout=20)
print("status:", r.status_code)

data = r.json()
print("top-level keys:", list(data.keys()))
jobs = data["jobs"]
print("jobs returned:", len(jobs), "| meta.total:", data["meta"]["total"])
print("first job:", jobs[0]["title"])
```

```python
for j in jobs[:8]:
    loc = (j.get("location") or {}).get("name", "")
    print(f"- {j['title']}  |  {loc}  |  {j['absolute_url']}")
```

```python
nl = [j for j in jobs if "amsterdam" in (j.get("location") or {}).get("name","").lower()]
data_roles = [j for j in jobs if any(k in j["title"].lower()
              for k in ["data", "machine learning", "ml", "scientist", "analytics"])]
print(f"Amsterdam jobs: {len(nl)} / {len(jobs)}")
print("data-ish titles:")
for j in data_roles[:10]:
    print("  -", j["title"], "|", (j.get("location") or {}).get("name",""))
```

```python
def fetch_board(token):
    u = f"https://boards-api.greenhouse.io/v1/boards/{token}/jobs"
    resp = requests.get(u, params={"content": "true"}, timeout=20)
    if resp.status_code != 200:
        print(f"  {token}: HTTP {resp.status_code} (wrong token?)")
        return []
    d = resp.json()
    print(f"  {token}: {len(d['jobs'])} jobs (meta.total {d['meta']['total']})")
    return d["jobs"]

for token in ["adyen", "databricks"]:   # try others; 404 just means wrong token
    fetch_board(token)
```

```python
adyen_jobs = fetch_board("adyen")
print("held", len(adyen_jobs), "raw Adyen jobs in memory")
```
